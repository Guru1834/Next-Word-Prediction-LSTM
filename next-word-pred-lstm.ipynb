```python
import numpy as np
import tensorflow as tf 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense,Input
import warnings 
warnings.filterwarnings('ignore')
```

```python
with open("/kaggle/input/datasets/ronikdedhia/next-word-prediction/1661-0.txt", "r", encoding="cp1252", errors="replace") as f:
    data = f.read()
```

# Tokenization 

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
len(tokenizer.word_index)
```

```python
input_seq = []
for sentence in data.split('\n'):
    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1,len(tokenized_sentence)):
        input_seq.append(tokenized_sentence[:i+1])
```

```python
max_len = max([len(x) for x in input_seq])
print(max_len)
```

# Padding

```python
padded_input_sequences = pad_sequences(input_seq, maxlen = max_len, padding= "pre")

X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]

y = to_categorical(y,num_classes =8932 )

y.shape
```

# LSTM Model

```python
model = Sequential([
    Input(shape=(20,)),
    Embedding(input_dim=8932, output_dim=400),
    LSTM(150),
    Dense(8932, activation='softmax')
])

model.summary()

model.compile(loss = "categorical_crossentropy",optimizer = 'adam',metrics = ['accuracy'])
```

```python
history = model.fit(X,y,epochs = 25)
```

# Prediction of Next Words

```python
sample_text = "sherlock"
for i in range(4):
    tokenized_text = tokenizer.texts_to_sequences([sample_text])[0]
    
    padded_text = pad_sequences([tokenized_text],maxlen = 20, padding= 'pre')
    
    nxt_word = np.argmax(model.predict(padded_text))
    
    for word, index in tokenizer.word_index.items():
        if index == nxt_word:
            sample_text = sample_text+" "+word
print(sample_text)
```
